```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
file_path = 'team.csv'
find_winning_column(file_path)
def find_team_wins_column(file_path):
    home_team_wins = 0
    away_team_wins = 0

    with open('Documents/match.csv', 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)

        for row in csv_reader:
            home_goals = int(row['home_team_goal'])
            away_goals = int(row['away_team_goal'])

            if home_goals > away_goals:
                home_team_wins += 1
            elif home_goals < away_goals:
                away_team_wins += 1

    if home_team_wins > away_team_wins:
        print(f"Home team has more wins: {home_team_wins} wins and away team with {away_team_wins} wins ")
    elif home_team_wins < away_team_wins:
        print(f"Away team has more wins: {away_team_wins} wins and home team with {home_team_wins} wins ")
    else:
        print("Both teams have an equal number of wins.")

    plt.pie([home_team_wins, away_team_wins], labels=['Home Team Wins', 'Away Team Wins'], colors=['yellow', 'blue'], autopct='%1.1f%%', startangle=140)
    plt.axis('equal') 
    plt.title('Comparison of Home Team Wins and Away Team Wins')
    plt.show()
```
