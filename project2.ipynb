```python
import pandas as pd
import matplotlib.pyplot as plt
credit_card = pd.read_csv('Downloads/credit_card (2).csv')
print(credit_card)

```

```python
credit_card.nunique()
```

```python
explode = (0.3, 0, 0)
total_df = credit_card['Reward']

totals = total_df.value_counts().tolist()
rewards = total_df.drop_duplicates().tolist()
_, ax1 = plt.subplots()
ax1.pie(totals, explode=explode, labels=rewards, autopct='%1.2f%%', startangle=360, colors=['green', 'red', 'blue'])
ax1.axis('equal') 
plt.title(' Most Popular Rewards')
plt.show()
```

```python
eplode = (0.1,0, 0)
income = credit_card['Income Level']
total_income = income.value_counts().list()
income= income.drop_duplicates().tolist()
__, ax2 = plt.subplots()
ax2.pie(total_income, eplode=eplode, labels=income , autopt='%1.2f%', startangle=360, colors=['purple', 'yellow', 'blue'])
ax2.axis('equal') 
plt.title(' Income level ')
plt.show()
```

```python
explode = (0.2,0,0)
mail = credit_card['Mailer Type']
total_ = mail.value_counts().tolist()
mail_= mail.drop_duplicates().tolist()
_, ax3 = plt.subplots()
ax3.pie(total_,  explode=explode , labels=mail , autopct='%1.2f%', startangle=360, colors=['orange', 'green', 'blue'])
ax3.axis('equal') 
plt.title(' Mailer Type ')
plt.show()
```
