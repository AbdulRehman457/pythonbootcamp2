```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("Documents/electronics.csv")
print (df)
```

```python
df['year'] = df['year'].astype(int)
df['brand'] = df['brand'].astype(str)
df['rating'] = df['rating'].astype(float)
df['item_id'] = df['item_id'].astype(int)
df['user_id'] = df['user_id'].astype(int)
df['timestamp'] = df['timestamp'].astype(str)
df['user_attr'] = df['user_attr'].astype(str)
df['model_attr'] = df['model_attr'].astype(str)
df['category'] = df['category'].astype(str)
```

```python
df.nunique()
```

```python
#used to drop duplicates values
rating = df['rating']
rating.dropna(inplace = True)
rating.drop_duplicates(inplace = True)
```

```python
df['year']= pd.DatetimeIndex(df['timestamp']).year
df.groupby('year')['rating'].count().plot(kind='bar')
```

```python
df['month']= pd.DatetimeIndex(df['timestamp']).month
df.groupby('month')['rating'].count().plot(kind='bar')
```
